```python
nama = "Dzaky Alam Fawaz Murtadha"
nim = "235150207111019"

# EP dan BVA
def validate_duration(duration):
  if 1 <= duration <= 4:
    return "DITERIMA"
  else:
      return "DITOLAK"
```

```python
# Decision Table
def can_borrow(is_registered, no_violation, room_available):
  if not is_registered:
    return "DITOLAK KARENA TIDAK TERDAFTAR"
  elif not no_violation:
    return "DITOLAK KARENA MEMILIKI PELANGGARAN"
  elif not room_available:
    return "DITOLAK KARENA RUANG TIDAK TERSEDIA"
  else:
    return "BOOKING BERHASIL"
```

```python
#State Transition
print(f"Tester:{nama} - {nim}")
print()

valid_transitions = {
    "AVAILABLE":{
        "book":"BOOKED"
    },
    "BOOKED":{
        "start":"IN_USE",
        "cancel":"CANCELLED"
    },
    "IN_USE":{
        "finish":"COMPLETED"
    },
    "COMPLETED":{},
    "CANCELLED":{}
}

def transition_state(current, event):
    if event in valid_transitions.get(current_state,{}):
      return valid_transitions[current_state][event]
    else:
      return "INVALID TRANSITION"

print()
print("TEST STATE TRANSITION VALID")

valid_state_test_cases = [
    ("AVAILABLE", "book"),
    ("BOOKED", "start"),
    ("IN_USE", "finish"),
    ("BOOKED", "cancel")
]

for current_state, event in  valid_state_test_cases:
  result = transition_state(current_state, event)
  print(f"{current_state} + {event} => {result}")

print()
print("TEST STATE TRANSITION INVALID")

invalid_state_test_cases = [
    ("AVAILABLE", "start"),
    ("AVAILABLE", "finish"),
    ("BOOKED", "book"),
    ("IN_USE", "book"),
    ("COMPLETED", "book"),
    ("CANCELLED", "start")
]

for current_state, event in invalid_state_test_cases:
  result = transition_state(current_state, event)
  print(f"{current_state} + {event} => {result}")
```
